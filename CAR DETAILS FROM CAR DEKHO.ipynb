```python
# Import All Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load data (replace with your actual data loading method)
data = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
data.head(5)
len(data)
```

```python

# Data preprocessing
data = data.fillna(method='ffill')  # Forward fill missing values

data.head(3)
```

```python
# 2. Encode categorical variables
data = pd.get_dummies(data, columns=['name', 'fuel', 'seller_type', 'transmission', 'owner'])
data
```

```python
# 3. Select features and target variable
X = data[['year', 'km_driven', 'seller_type_Individual' ,'transmission_Manual',	'owner_First Owner',	'owner_Fourth & Above Owner',	'owner_Second Owner',	'owner_Test Drive Car',	'owner_Third Owner']] 
y = data['selling_price']

```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
len(X_train)
```

```python
len(X_test)
X_test.head(5)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

```

```python
# Make predictions
y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

```

```python

rmse = mse ** 0.5
print("Root Mean Squared Error:", rmse)

```

```python
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```
